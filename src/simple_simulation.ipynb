```python
from collections import defaultdict
from world_cup_simulation import simulate_world_cup, dummy_skill_function

N = 1000

# dict of lists to store results for each team
all_results = defaultdict(list)

for _ in range(N):
    placements = simulate_world_cup(sf=dummy_skill_function, verbose=False)
    for team, stage in placements.items():
        all_results[team].append(stage)
```

```python
import matplotlib.pyplot as plt
from collections import Counter

# count placements for austria
austria_counts = Counter(all_results['austria'])

# ensure consistent order
stages = ['GROUPS', 'R32', 'R16', 'QF', 'SF', 'F', 'WINNER']
counts = [austria_counts.get(stage, 0) for stage in stages]

plt.figure(figsize=(8,5))
plt.bar(stages, counts, color='red')
plt.title("Austria - World Cup Placement Distribution (1000 simulations)")
plt.ylabel("Count")
plt.xlabel("Stage Reached")
plt.show()
```
